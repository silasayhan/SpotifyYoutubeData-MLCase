```python
# SÃ¼tun AdlarÄ±nÄ± SadeleÅŸtir: BazÄ± sÃ¼tunlar fazla boÅŸluk iÃ§erebilir veya karmaÅŸÄ±k isimlere sahip olabilir. 
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
df.columns
```

```python
# Kategorik TÃ¼r DÃ¶nÃ¼ÅŸÃ¼mÃ¼
cat_cols = ['album_type', 'channel', 'artist', 'album']
for col in cat_cols:
    df[col] = df[col].astype('category')
```

```python
# KÃ¼Ã§Ã¼k eksikleri olan sayÄ±sal sÃ¼tunlar iÃ§in satÄ±rlarÄ± silelim.
numeric_subset = ['danceability', 'energy', 'key', 'loudness', 'speechiness',
                  'acousticness', 'instrumentalness', 'liveness',
                  'valence', 'tempo', 'duration_ms']
df.dropna(subset = numeric_subset, inplace = True)

# YouTube ile ilgili sÃ¼tunlarda eksik varsa, sil.
df.dropna(subset = ['url_youtube', 'title', 'channel', 'views'], inplace = True)

# MantÄ±ksal sÃ¼tunlarÄ± bool tipine Ã§evir ve eksikleri 'False' yap.
df['licensed'] = df['licensed'].map({'True': True, 'False': False})
df['official_video'] = df['official_video'].map({'True': True, 'False': False})
df['licensed'].fillna(False, inplace = True)
df['official_video'].fillna(False, inplace = True)

# Likes, Comments, Stream, ortanca ile doldur.
df['likes'].fillna(df['likes'].median(), inplace = True)
df['comments'].fillna(df['comments'].median(), inplace = True)
df['stream'].fillna(df['stream'].median(), inplace = True)

# Ä°stenmeyen SÃ¼tunlarÄ± Silme
df.drop(columns = ['unnamed:_0', 'uri', 'url_spotify', 'url_youtube', 'description'], inplace = True)
```

```python
# Kategorik SÃ¼tunlarÄ±n Benzersiz DeÄŸer SayÄ±sÄ±
output_data = []
for col in df.columns:
    if df.loc[:, col].nunique() <= 5:
        unique_values = df.loc[:, col].unique()
        output_data.append([col, df.loc[:, col].nunique(), unique_values, df.loc[:, col].dtype])
    else:
        output_data.append([col, df.loc[:, col].nunique(),"-", df.loc[:, col].dtype])
output_df = pd.DataFrame(output_data, columns=['Column Name', 'Number of Unique Values', ' Unique Values ', 'Data Type'])
(output_df)
```

```python
# Q1: YouTube'da en yÃ¼ksek izlenme sayÄ±sÄ±na sahip ilk 10 sanatÃ§Ä±yÄ± bul.
top_artists = df.groupby('artist')['views'].sum().sort_values(ascending = False).head(10)
print(top_artists.reset_index())
```

```python
# Q2: Spotify'da en Ã§ok dinlenen 10 parÃ§ayÄ± tablo olarak listeleyeceÄŸiz.
top_tracks = df[['title', 'artist', 'stream']].sort_values(by = 'stream', ascending = False).head(10)
print(top_tracks.reset_index(drop = True))

# Q2A: Spotify'da en az dinlenen parÃ§alarÄ± listeleyeceÄŸiz.
x = df[['title', 'stream']]
least_stream_track = x.sort_values(by = ['stream'], ascending = True).head(5)
least_stream_track
```

```python
# Q3: Spotify'daki en yaygÄ±n albÃ¼m tÃ¼rleri nelerdir? Her bir tÃ¼re kaÃ§ parÃ§a dÃ¼ÅŸÃ¼yor?
df['album_type'].value_counts()
```

```python
# Q4: FarklÄ± albÃ¼m tiplerine gÃ¶re ortalama gÃ¶rÃ¼ntÃ¼lenme, beÄŸeni ve yorum sayÄ±sÄ±nÄ± nasÄ±l karÅŸÄ±laÅŸtÄ±rÄ±rsÄ±n?
df_melted = pd.melt(df, 
                    id_vars='album_type', 
                    value_vars=['views', 'likes', 'comments'],
                    var_name='Attribute', 
                    value_name='Total')

plt.figure(figsize = (5, 3))
sns.barplot(data = df_melted, x = 'album_type', y = 'Total', hue = 'Attribute')
plt.title('Album Type BazÄ±nda Ortalama Views, Likes, Comments KarÅŸÄ±laÅŸtÄ±rmasÄ±')
plt.xlabel('Album Type')
plt.ylabel('Ortalama DeÄŸer')
plt.xticks(rotation = 45)
plt.show()
```

```python
# Q5: En yÃ¼ksek gÃ¶rÃ¼ntÃ¼lenme sayÄ±sÄ±na sahip ilk 5 YouTube kanalÄ±nÄ± listele.
top_channels = df.groupby('channel')['views'].sum().sort_values(ascending=False).head(5).reset_index()
print(top_channels)

```

```python
# Q6: En yÃ¼ksek gÃ¶rÃ¼ntÃ¼lenmeye sahip parÃ§a hangisidir?
top_track = df.loc[df['views'].idxmax()]

# ParÃ§anÄ±n ismini, sanatÃ§Ä±sÄ±nÄ± ve gÃ¶rÃ¼ntÃ¼lenme sayÄ±sÄ±nÄ± yazdÄ±r.
print(f"En YÃ¼ksek GÃ¶rÃ¼ntÃ¼lenmeye Sahip ParÃ§a: {top_track['track']}")
print(f"SanatÃ§Ä±: {top_track['artist']}")
print(f"GÃ¶rÃ¼ntÃ¼lenme SayÄ±sÄ±: {top_track['views']}")
```

```python
# Q7: YouTube verilerinde bir ÅŸarkÄ±nÄ±n izlenme sayÄ±sÄ±na gÃ¶re beÄŸeni yÃ¼zdesi en yÃ¼ksek olan ilk 7 ÅŸarkÄ±yÄ± bul.

# Like-View OranÄ± 
df['like_to_view_ratio'] = df['likes'] / df['views']

# Views = 0 olan satÄ±rlarÄ± filtrelemek 0!a bÃ¶lme hatasÄ±nÄ± engeller.
df = df[df['views'] > 0]

top7_ratio = df.sort_values(by = 'like_to_view_ratio', ascending = False).head(7)

top7_ratio[['track', 'artist', 'views', 'likes', 'like_to_view_ratio']]

plt.figure(figsize = (5, 3))
sns.barplot(x = 'like_to_view_ratio', y = 'track', data = top7_ratio, palette = 'viridis')
plt.title('Top 7 Tracks by Like-to-View Ratio')
plt.xlabel('Like-to-View Ratio')
plt.ylabel('Track Name')
plt.show()

```

```python
# Q8: Veri setindeki en hareketli ÅŸarkÄ±larÄ± listele.

#  ParÃ§alarÄ± dans edilebilirlik deÄŸerine gÃ¶re bÃ¼yÃ¼kten kÃ¼Ã§Ã¼ÄŸe sÄ±ralar.
top_dance_tracks = df.sort_values(by = 'danceability', ascending = False).head(10)

top_dance_albums = top_dance_tracks[['album', 'track', 'danceability']]

top_dance_albums
```

```python
""" Q9: YouTube ve Spotifyâ€™daki dÃ¶rt temel sayÄ±sal metrik arasÄ±ndaki iliÅŸkiyi inceleyeceÄŸiz:
    Views (YouTube Ä°zlenme SayÄ±sÄ±)
    Likes
    Comments
    Stream (Spotify Dinlenme SayÄ±sÄ±)
    Bunlar arasÄ±ndaki korelasyonu bulacaÄŸÄ±z. Korelasyon, iki deÄŸiÅŸken arasÄ±ndaki iliÅŸkinin gÃ¼cÃ¼nÃ¼ ve yÃ¶nÃ¼nÃ¼ Ã¶lÃ§er. 
    Pearson korelasyon katsayÄ±sÄ± -1 ile 1 arasÄ±nda deÄŸer alÄ±r:
    1 â†’ Tam Pozitif Ä°liÅŸki
    0 â†’ Ä°liÅŸki Yok
    -1 â†’ Tam Negatif Ä°liÅŸki"""

# SayÄ±sal SÃ¼tunlardan Korelasyon Matrisi OluÅŸturma
num_cols = ['views', 'likes', 'comments', 'stream']
corr_matrix = df[num_cols].corr()

# Korelasyon Matrisini GÃ¶rselleÅŸtirme
plt.figure(figsize = (4, 3))
sns.heatmap(corr_matrix, annot = True, cmap = 'coolwarm', fmt = ".2f")
plt.title('Ä°zlenme SayÄ±sÄ±, BeÄŸeni, Yorum ve Dinlenme SayÄ±sÄ± ArasÄ±ndaki Korelasyon Matrisi')
plt.show()

```
