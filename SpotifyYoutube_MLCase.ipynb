```python
# Sütun Adlarını Sadeleştir: Bazı sütunlar fazla boşluk içerebilir veya karmaşık isimlere sahip olabilir. 
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
df.columns
```

```python
# Kategorik Tür Dönüşümü
cat_cols = ['album_type', 'channel', 'artist', 'album']
for col in cat_cols:
    df[col] = df[col].astype('category')
```

```python
# Küçük eksikleri olan sayısal sütunlar için satırları silelim.
numeric_subset = ['danceability', 'energy', 'key', 'loudness', 'speechiness',
                  'acousticness', 'instrumentalness', 'liveness',
                  'valence', 'tempo', 'duration_ms']
df.dropna(subset = numeric_subset, inplace = True)

# YouTube ile ilgili sütunlarda eksik varsa, sil.
df.dropna(subset = ['url_youtube', 'title', 'channel', 'views'], inplace = True)

# Mantıksal sütunları bool tipine çevir ve eksikleri 'False' yap.
df['licensed'] = df['licensed'].map({'True': True, 'False': False})
df['official_video'] = df['official_video'].map({'True': True, 'False': False})
df['licensed'].fillna(False, inplace = True)
df['official_video'].fillna(False, inplace = True)

# Likes, Comments, Stream, ortanca ile doldur.
df['likes'].fillna(df['likes'].median(), inplace = True)
df['comments'].fillna(df['comments'].median(), inplace = True)
df['stream'].fillna(df['stream'].median(), inplace = True)

# İstenmeyen Sütunları Silme
df.drop(columns = ['unnamed:_0', 'uri', 'url_spotify', 'url_youtube', 'description'], inplace = True)
```

```python
# Kategorik Sütunların Benzersiz Değer Sayısı
output_data = []
for col in df.columns:
    if df.loc[:, col].nunique() <= 5:
        unique_values = df.loc[:, col].unique()
        output_data.append([col, df.loc[:, col].nunique(), unique_values, df.loc[:, col].dtype])
    else:
        output_data.append([col, df.loc[:, col].nunique(),"-", df.loc[:, col].dtype])
output_df = pd.DataFrame(output_data, columns=['Column Name', 'Number of Unique Values', ' Unique Values ', 'Data Type'])
(output_df)
```

```python
# Q1: YouTube'da en yüksek izlenme sayısına sahip ilk 10 sanatçıyı bul.
top_artists = df.groupby('artist')['views'].sum().sort_values(ascending = False).head(10)
print(top_artists.reset_index())
```

```python
# Q2: Spotify'da en çok dinlenen 10 parçayı tablo olarak listeleyeceğiz.
top_tracks = df[['title', 'artist', 'stream']].sort_values(by = 'stream', ascending = False).head(10)
print(top_tracks.reset_index(drop = True))

# Q2A: Spotify'da en az dinlenen parçaları listeleyeceğiz.
x = df[['title', 'stream']]
least_stream_track = x.sort_values(by = ['stream'], ascending = True).head(5)
least_stream_track
```

```python
# Q3: Spotify'daki en yaygın albüm türleri nelerdir? Her bir türe kaç parça düşüyor?
df['album_type'].value_counts()
```

```python
# Q4: Farklı albüm tiplerine göre ortalama görüntülenme, beğeni ve yorum sayısını nasıl karşılaştırırsın?
df_melted = pd.melt(df, 
                    id_vars='album_type', 
                    value_vars=['views', 'likes', 'comments'],
                    var_name='Attribute', 
                    value_name='Total')

plt.figure(figsize = (5, 3))
sns.barplot(data = df_melted, x = 'album_type', y = 'Total', hue = 'Attribute')
plt.title('Album Type Bazında Ortalama Views, Likes, Comments Karşılaştırması')
plt.xlabel('Album Type')
plt.ylabel('Ortalama Değer')
plt.xticks(rotation = 45)
plt.show()
```

```python
# Q5: En yüksek görüntülenme sayısına sahip ilk 5 YouTube kanalını listele.
top_channels = df.groupby('channel')['views'].sum().sort_values(ascending=False).head(5).reset_index()
print(top_channels)

```

```python
# Q6: En yüksek görüntülenmeye sahip parça hangisidir?
top_track = df.loc[df['views'].idxmax()]

# Parçanın ismini, sanatçısını ve görüntülenme sayısını yazdır.
print(f"En Yüksek Görüntülenmeye Sahip Parça: {top_track['track']}")
print(f"Sanatçı: {top_track['artist']}")
print(f"Görüntülenme Sayısı: {top_track['views']}")
```

```python
# Q7: YouTube verilerinde bir şarkının izlenme sayısına göre beğeni yüzdesi en yüksek olan ilk 7 şarkıyı bul.

# Like-View Oranı 
df['like_to_view_ratio'] = df['likes'] / df['views']

# Views = 0 olan satırları filtrelemek 0!a bölme hatasını engeller.
df = df[df['views'] > 0]

top7_ratio = df.sort_values(by = 'like_to_view_ratio', ascending = False).head(7)

top7_ratio[['track', 'artist', 'views', 'likes', 'like_to_view_ratio']]

plt.figure(figsize = (5, 3))
sns.barplot(x = 'like_to_view_ratio', y = 'track', data = top7_ratio, palette = 'viridis')
plt.title('Top 7 Tracks by Like-to-View Ratio')
plt.xlabel('Like-to-View Ratio')
plt.ylabel('Track Name')
plt.show()

```

```python
# Q8: Veri setindeki en hareketli şarkıları listele.

#  Parçaları dans edilebilirlik değerine göre büyükten küçüğe sıralar.
top_dance_tracks = df.sort_values(by = 'danceability', ascending = False).head(10)

top_dance_albums = top_dance_tracks[['album', 'track', 'danceability']]

top_dance_albums
```

```python
""" Q9: YouTube ve Spotify’daki dört temel sayısal metrik arasındaki ilişkiyi inceleyeceğiz:
    Views (YouTube İzlenme Sayısı)
    Likes
    Comments
    Stream (Spotify Dinlenme Sayısı)
    Bunlar arasındaki korelasyonu bulacağız. Korelasyon, iki değişken arasındaki ilişkinin gücünü ve yönünü ölçer. 
    Pearson korelasyon katsayısı -1 ile 1 arasında değer alır:
    1 → Tam Pozitif İlişki
    0 → İlişki Yok
    -1 → Tam Negatif İlişki"""

# Sayısal Sütunlardan Korelasyon Matrisi Oluşturma
num_cols = ['views', 'likes', 'comments', 'stream']
corr_matrix = df[num_cols].corr()

# Korelasyon Matrisini Görselleştirme
plt.figure(figsize = (4, 3))
sns.heatmap(corr_matrix, annot = True, cmap = 'coolwarm', fmt = ".2f")
plt.title('İzlenme Sayısı, Beğeni, Yorum ve Dinlenme Sayısı Arasındaki Korelasyon Matrisi')
plt.show()

```
